```
import numpy as np
import tensorflow as tf
```

```
# 실행할 떄마다 같은 결과를 출력하기 위한 seed 값 설정
seed = 2021
np.random.seed(seed)
tf.random.set_seed(seed)
```

```
# 준비된 수술 환자 데이터를 불러들임
from google.colab import files
uploaded = files.upload()
uploaded
```

```
filename = list(uploaded.keys())[0]
filename
```

```
Data_set = np.loadtxt(filename, delimiter=",")
Data_set[:3, :]
```

### 데이터 정규화

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(Data_set[:, :-1])
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, Data_set[:, -1], stratify=Data_set[:, -1], random_state=seed
)
X_train.shape, X_test.shape
```

### 모델 정의/설정/학습/평가

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```
model = Sequential()
model.add(Dense(24, input_shape=(17,), activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```
history = model.fit(X_train, y_train, validation_split=0.2, epochs=30, batch_size=10)
```

```
acc = model.evaluate(X_test, y_test)
print(f'Accuracy: {acc[1]:.4f}')
```
